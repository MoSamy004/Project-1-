```python
# our data clean .. so lets make some new columns
```

```python
import numpy as np

# Create age category column using numpy.where for vectorized conditional assignment
df['age category'] = np.where((df['Age'] >= 18) & (df['Age'] <= 30), 'youth',
                     np.where((df['Age'] >= 31) & (df['Age'] <= 40), 'middle age',
                     np.where((df['Age'] >= 41) & (df['Age'] <= 50), 'senior', 'old age')))

df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df.drop_duplicates()
```

```python
df['Client category'] = np.where((df['Price per Unit'] >= 25) & (df['Price per Unit'] <= 30), 'Normal Client',
                     np.where((df['Price per Unit'] >= 31) & (df['Price per Unit'] <= 50), 'Loyal Client',
                     np.where((df['Price per Unit'] >= 51) & (df['Price per Unit'] <= 70), 'Gold Client',
                     np.where((df['Price per Unit'] >= 71) & (df['Price per Unit'] <= 80), 'Platinum Client',
                    np.where((df['Price per Unit'] >= 81), 'VIP Client', 'Unknown')))))
df.head()
```

```python
df.to_csv(r'E:\Oasis Infobyte Internship\Project 1\Data\Cleand_data_set_2.csv', index=False)
```
